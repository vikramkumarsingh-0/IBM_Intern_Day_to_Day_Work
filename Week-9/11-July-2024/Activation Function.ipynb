```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Activation function

```python
plt.figure(figsize = (4, 4))
```

```python
x = np.linspace(-5,5)
```

```python
x
```

```python
# sigmoid function
```

```python
def sigm(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x, sigm(x))
plt.title("sigmoid activation function")
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

plt.plot(x, tanh(x))
plt.title("tanH activation function")
plt.show()
```

```python
def relu(x):
    list = []
    for i in x:
        if i>=0:
            list.append(i)
        else:
            list.append(0)
    return list

plt.plot(x, relu(x))
plt.title("Rectified linear Unit activation function")
plt.show()
```

```python
def lrelu(x):
    list = []
    for i in x:
        if (i>=0):
            list.append(i)
        else:
            list.append(0.01*i)
    return list

plt.plot(x, lrelu(x))
plt.title("leaky Rectified linear Unit activation function")
plt.show()
```

```python
def softm(x):
    return (np.exp(x)/ np.sum(np.exp(x)))

plt.plot(x, softm(x))
plt.title("softmax activation function")
plt.show()
```
