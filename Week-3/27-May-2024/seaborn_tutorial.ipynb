```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
x = df['total_bill']
y = df['tip']
```

```python
sns.scatterplot(x='total_bill',y='tip', data=df, hue='sex')
```

```python
import matplotlib.pyplot as plt
sns.relplot(
data=df,
x = 'total_bill', y = 'tip',
hue = 'time',
size = 'size',
sizes=(10, 100),
palette = 'Accent_r'
)

plt.xlabel('Total Bill in USD $')
plt.ylabel('Yips to the service in usd $')
plt.title('New York Restaurent Data')
```

```python
sns.lineplot(data=df, x='total_bill', y='size', hue='sex', style='sex', markers=True, palette='Blues_d', dashes=True)

plt.xlabel('Total Bill in USD $')
plt.ylabel('Yips to the service in usd $')
plt.title('New York Restaurent Data')
```
