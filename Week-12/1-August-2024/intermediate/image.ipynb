```python
os.listdir(path)
```

```python
from tensorflow import keras
```

```python
train = 'seg_train/seg_train'
test = 'seg_test/seg_test'
pred = 'seg_pred/seg_pred'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# normalisation
data = ImageDataGenerator(rescale = 255/255)
train_set = data.flow_from_directory(
    train,
    classes = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street'],
    class_mode = 'categorical',
    target_size = (100, 100) # size of image
)
```

```python
# normalisation
data = ImageDataGenerator(rescale = 255/255)
test_set = data.flow_from_directory(
    test,
    classes = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street'],
    class_mode = 'categorical',
    target_size = (100, 100),
)
```

```python
x_train, y_train = next(train_set)
```

```python
x_test, y_test = next(test_set)
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D
```

```python
model = Sequential()
```

```python
model.add(Conv2D(32, (3,3), input_shape = (100, 100, 3), activation = 'relu'))
```

```python
model.add(MaxPool2D(pool_size = (2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(100, activation='relu'))
```

```python
model.add(Dense(6, activation= 'sigmoid'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
Model = model.fit(x_train, y_train, epochs =10)
```

```python
from  sklearn.metrics import accuracy_score
```

```python
score = model.evaluate(x_test, y_test)
```

```python
score[1]
```

```python
# from sklearn import preprocesing, MinMaxScaler
loss = score[0]
accuracy = score[1]
```
